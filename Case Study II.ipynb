```python
import pandas as pd
import seaborn as sns
```

### List Comprehension

```python
# Görev 1
# List comprehension yapısı kullanarak car_crashes verisindeki numeric değişkenlerin isimlerini büyük harfe çeviriniz
# ve başına NUM ekleyiniz.

df = sns.load_dataset('car_crashes')
['NUM_' + col.upper() if df[col].dtype != 'O' else col.upper() for col in df.columns]
```

```python
# Görev 2
# List comprehension yapısı kullanarak car_crashes verisinde isminde "no" barındırmayan değişkenlerin isimlerinin
# sonuna FLAG yazınız.
df.head()
[(col.upper() + '_FLAG') if ('no' not in col) else col.upper() for col in df.columns]
```

```python
# Görev 3
# List comprehension yapısı kullanarak aşağıda verilen değişken isimlerinden farklı olan değişkenlerin isimlerini seçiniz ve
# yeni bir dataframe oluşturunuz.

og_list=['abbrev', 'no_previous']
new_cols = [col for col in df.columns if (col not in og_list)]
new_df = df[new_cols]; new_df.head()
```

### Pandas Alıştırmalar

```python
# Görev 1
# Seaborn kütüphanesi içerisinden Titanic veri setini tanımlayınız.

titanic = sns.load_dataset('titanic'); titanic.head()
```

```python
# Görev 2
# Titanic veri setindeki kadın ve erkek yolcuların sayısını bulunuz.

titanic['sex'].value_counts().to_frame('count')
```

```python
# Görev 3
# Her bir sütuna ait unique değerlerin sayısını bulunuz.

titanic.nunique()
```

```python
# Görev 4
# pclass değişkeninin unique değerlerinin sayısını bulunuz.
titanic.pclass.nunique()
```

```python
# Görev 5
# pclass ve parch değişkenlerinin unique değelerinin sayısını bulunuz.

titanic[['pclass', 'parch']].nunique()
```

```python
# Görev 6
# embarked değişkeninin tipini kontrol ediniz. Tipini category olarak değiştiriniz ve tekrar kontrol ediniz.

titanic.embarked.dtype
```

```python
titanic.embarked = titanic.embarked.astype('category')
```

```python
titanic.embarked.dtype
```

```python
# Görev 7
# embarked değeri C olanların tüm bilgilerini gösteriniz.

titanic[titanic.embarked == 'C'].head()
```

```python
# Görev 8
# embarked değeri S olmayanların tüm bilgilerini gösteriniz.

titanic[titanic.embarked != 'S'].head()
```

```python
# Görev 9
# Yaşı 30'dan küçük ve kadın olan yolcuların tüm bilgilerini gösteriniz

titanic[(titanic['age'] < 30) & (titanic['sex'] == 'female')].head()
```

```python
# Görev 10
# Fare'i 500'den büyük veya yaşı 70'den büyük yolcuların bilgilerini gösteriniz.

titanic[(titanic['fare'] > 500) | (titanic['age'] > 70)].head()
```

```python
# Görev 11
# Her bir değişkendeki boş değerlerin sayısını bulunuz.

titanic.isnull().sum()
```

```python
# Görev 12
# who değişkenini dataframe'den çıkarınız.
titanic
titanic.drop('who', axis=1, inplace=True); titanic.head()
```

```python
# Görev 13
# deck değişkenindeki boş değerleri deck değişkeninin en çok tekrar eden değeri (mode) ile doldurunuz.

titanic['deck'].fillna(titanic['deck'].mode()[0])
```

```python
# Görev 14
# age değişkenindeki boş değerleri age değişkeninin medyanı ile doldurunuz.

titanic['age'].fillna(titanic['age'].median())
```

```python
# Görev 15
# survived değişkeninin pclass ve cinsiyet değişkenleri kırılımında sum, count, mean değerlerini bulunuz.

titanic.groupby(['pclass', 'sex']).agg({'survived' : ['sum', 'count', 'mean']})
```

```python
# Görev 16
# 30 yaşın altında olanlar 1, 30'a eşit ve üstünde olanlara 0 verecek bir fonksiyon yazın. Yazdığınız fonksiyonu kullanarak
# titanic veri setinde age_flag adında bir değişken oluşturunuz (apply ve lambda yapılarını kullanınız).

titanic['age_flag'] = titanic['age'].apply(lambda x: 1 if x >= 30 else 0)
```

```python
# Görev 17
# seaborn kütüphanesi içerisinden Tips veri setini tanımlayınız

tips = sns.load_dataset('tips'); tips.head()
```

```python
# Görev 18
# Time değişkeninin kategorilerine (Dinner, Lunch) göre total_bill değerlerinin toplamını min, max ve ortalamasını bulunuz.

tips.groupby('time').agg({'total_bill' : ['min', 'max', 'mean']})
```

```python
# Görev 19
# Günlere ve time'a göre total_bill değerlerinin toplamını, min, max ve ortalamasını bulunuz.

tips.groupby(['day', 'time']).agg({'total_bill' : ['min', 'max', 'mean']})
```

```python
# Görev 20
# Lunch zamanına göre ve kadın müşterilere ait total_bill ve tip değerlerinin day'e göre toplamını,
# min, max ve ortalamasını bulunuz.

tips[(['sex'] == 'Female') & (tips['time'] == 'Lunch')].groupby('day').agg({
    'total_bill': ['sum', 'min', 'max'],
    'tip': ['sum', 'min', 'max']
})
```

```python
# Görev 21
# Size'ı 3'ten küçük, total_bill'i 10'dan büyük olan siparişlerin ortalaması nedir? (loc kullanınız)

tips.loc[(tips['size'] < 3) & (tips['total_bill'] > 10), ['total_bill']].mean()
```

```python
tips.head()
```

```python
# Görev 22
# total_bill_tip_sum adında yeni bir değişken oluşturunuz. Her bir müşterinin ödediği total_bill ve tip'in toplamını versin.

tips['total_bill_tip_sum'] = tips['total_bill'] + tips['tip']
```

```python
# Görev 23
# total_bill_tip_sum değişkenine göre büyükten küçüğe sıralayınız ve ilk 30 kişiyi yeni bir dataframe'e atayınız.
tips_30 = tips.sort_values(by='total_bill_tip_sum', ascending = False).head(30)
```
