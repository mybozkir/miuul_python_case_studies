## Görev 4: City-Concept-Season kırılımının çıktısını PRICE'a göre sıralayınız.

#### Elde ettiğiniz çıktıyı agg_df olarak kaydediniz.

```python
#Şehir-Concept-Sezon
agg_df = gezinomi.groupby(['SaleCityName', 'ConceptName', 'Seasons'])[['Price']].mean().sort_values(by=['Price'], ascending=False);
agg_df.head()
```

## Görev 5: Index'te yer alan isimleri değişken ismine çeviriniz.

```python
agg_df = agg_df.reset_index(); agg_df.head()
```

## Görev 6: Yeni seviye tabanlı müşterileri (persona) tanımlayınız.

#### Yeni seviye tabanlı satışları tanımlayınız ve veri setine değişken olarak ekleyiniz.
#### Yeni eklenecek değişkenin adı: sales_level_based
#### Önceki soruda elde edeceğiniz çıktıdaki gözlemleri biraraya getirerek sales_lavel_based değişkenini oluşturmanız gerek.

```python
agg_df['sales_level_based'] = agg_df['SaleCityName'].str.upper() + '_' + agg_df['ConceptName'].str.upper() + '_' + agg_df['Seasons'].str.upper()
```

```python
agg_df.head()
```

## Görev 7: Yeni müşterileri (personaları) segmentlere ayırınız.

#### Yeni personaları PRICE'a göre 4 segmente ayırınız.
#### Segmentleri SEGMENT isimlendirmesi ile değişken olarak agg_df'e ekleyiniz. 
#### Segmentleri betimleyiniz (Segmentlere göre group by yapıp price mean, max, sum'larını alınız.)

```python
gezinomi['Segment'] = pd.qcut(gezinomi['Price'], 4, labels=['D', 'C', 'B', 'A']); gezinomi.head()
```

```python
gezinomi.groupby('Segment').agg({
    'Price': ['mean', 'max', 'sum']
})
```
